```python
import itertools  
```

```python
data = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
```

```python
with open("d1.txt", "r") as f:
    data = f.read()
```

```python
[
    v % 100
    for v in list(
        itertools.accumulate(
            [50]
            + [int(v[1:]) if v[0] == "R" else -int(v[1:]) for v in data.split("\n")]
        )
    )
].count(0)
```

```python
sum(
    [
        1 if v % 100 == 0 else 0
        for v in list(
            itertools.accumulate(
                [50]
                + list(
                    itertools.chain(
                        *[
                            [1] * int(v[1:]) if v[0] == "R" else [-1] * int(v[1:])
                            for v in data.split("\n")
                        ]
                    )
                )
            )
        )
    ]
)
```
