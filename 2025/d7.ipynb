```python
manif = data.split("\n")
```

```python
tach = np.zeros((len(manif), len(manif[0])))
tach[0, list(manif[0]).index("S")] = 1
tach
```

```python
def one_step(tach, manif, splits1, splits2):
    new_tach = tach.copy()
    for x in range(0, tach.shape[1], 1):
        for y in range(tach.shape[0]-1):
            if (tach[y, x]>0) & (manif[y+1][x]=="^"):
                new_tach[y+1, x-1] = new_tach[y+1, x-1] + new_tach[y, x]
                new_tach[y+1, x+1] = new_tach[y+1, x+1] + new_tach[y, x]
                splits1 += 1
                splits2 += new_tach[y, x]
                new_tach[y, x] = 0
            elif (tach[y, x]>0) & (manif[y+1][x]=="."):
                new_tach[y+1, x] = new_tach[y+1, x] + new_tach[y, x]
                new_tach[y, x] = 0
    return new_tach, splits1, splits2
```

```python
splits1 = 0
splits2 = 1
new_tach = tach.copy()
for y in range(tach.shape[0]-1):
    new_tach, splits1, splits2 = one_step(new_tach, manif, splits1, splits2)
```
