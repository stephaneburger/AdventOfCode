```python
with open("d3.txt", "r") as f:
    data = f.read()
```

```python
banks = [[int(v2) for v2 in list(v)] for v in data.split("\n")]
# banks
```

```python
joltage1 = 0
joltage2 = 0
for bank in banks:
    # Part 1
    d1 = max(bank[:-1])
    i = bank.index(d1)
    d2 = max(bank[i+1:])
    joltage1 += int(str(d1)+str(d2))

    # Part 2
    i = -1
    l = []
    for n in range(1, 13, 1):
        e = -12+n
        if e !=0:
            d = max(bank[i+1 : -12+n])
            i = i + 1 + bank[i+1 : -12+n].index(d)
        else:
            d = max(bank[i+1 :])
            i = i + 1 + bank[i+1 :].index(d)
        l.append(str(d))
    joltage2 += int("".join(l))
joltage1, joltage2
```
