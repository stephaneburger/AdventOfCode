```python
def paths_start_end(start, end, connections=connections, extra_iterations=8):
    paths = {start:1}
    new_paths = {}
    done = False
    more_it = 0
    
    while True:
        new_paths = {}
        for k, v in paths.items():
            if k == end:
                if k in new_paths.keys():
                    new_paths[k] = new_paths[k] + v
                else:
                    new_paths[k] = v
                continue
            new_heads = connections[k]
            for n in new_heads:
                if n in new_paths.keys():
                    new_paths[n] = new_paths[n] + v
                else:
                    new_paths[n] = v
                if n == end:
                    done = True
        paths = new_paths
        # print(paths)
        if done:
            more_it += 1
        if done & (more_it>extra_iterations):
            break
    return paths[end]
```
