```python
with open("d8.txt", "r") as f:
    data = f.read()
its = 1000
```

```python
junctions = [[int(v) for v in l.split(',')] for l in data.split("\n")]
```

```python
dist_matrix = pd.DataFrame(columns = range(len(junctions)), index = range(len(junctions)))
for i, j1 in enumerate(junctions):
    for j, j2 in enumerate(junctions):
        if i == j:
            dist_matrix.loc[i, j] = 99999999999999
        else:
            dist_matrix.loc[i, j] = sqrt((j1[0]-j2[0])**2+(j1[1]-j2[1])**2+(j1[2]-j2[2])**2)
```

```python
def find_closest_pair(dist_matrix):
    # min_dists = dist_matrix.idxmin().reset_index()
    # min_dists.columns = ["j1", "j2"]
    # min_dists["dist"] = min_dists.apply(lambda row: dist_matrix.loc[row['j1'], row['j2']],  axis=1)
    # pair = sorted(min_dists.loc[min_dists["dist"].idxmin(), ["j1", "j2"]].astype(int).tolist())
    # return pair
    return [int(v) for v in list(np.unravel_index(dist_matrix.to_numpy().argmin(), dist_matrix.shape))]

def recombine_cluster(clusters):
    sets = list(clusters)
    merged = True
    while merged:
        merged = False
        result = []
        while sets:
            first, *rest = sets
            first = set(first)
            sets = []
            for s in rest:
                if first.intersection(s):
                    first |= s
                    merged = True
                else:
                    sets.append(s)
            result.append(first)
        sets = result
    return sets
```

```python
dist_matrix_copy = dist_matrix.copy()
clusters = []
for n in range(its):
    pair = find_closest_pair(dist_matrix_copy)
    dist_matrix_copy.loc[pair[0], pair[1]] = 99999999999999
    dist_matrix_copy.loc[pair[1], pair[0]] = 99999999999999
    if len(clusters) == 0:
        clusters.append(set(pair))
    else:
        paired = False
        for cluster in clusters:
            if (pair[0] in cluster) | (pair[1] in cluster):
                paired = True
                cluster.add(pair[0])
                cluster.add(pair[1])
        if not paired:
            clusters.append(set(pair))
clusters = recombine_cluster(clusters)
```
