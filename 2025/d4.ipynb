```python
with open("d4.txt", "r") as f:
    data = f.read()
```

```python
rolls_plan = [list(l) for l in data.split("\n")]
rolls_plan = [['.'] * (len(rolls_plan[0])+2)] + [['.']+l+['.'] for l in rolls_plan] + [['.'] * (len(rolls_plan[0])+2)]
```

```python
counter = 0
for x in range(1, len(rolls_plan[0])-1):
    for y in range( 1, len(rolls_plan)-1):
        if ((rolls_plan[y-1][x-1:x+2].count("@") + rolls_plan[y][x-1:x+2].count("@") + rolls_plan[y+1][x-1:x+2].count("@"))<=4)&(rolls_plan[y][x]=="@"):
            counter +=1
counter
```

```python
counter = 0
prev_counter = -1
while prev_counter < counter:
    prev_counter = counter
    for x in range(1, len(rolls_plan[0])-1):
        for y in range( 1, len(rolls_plan)-1):
            if ((rolls_plan[y-1][x-1:x+2].count("@") + rolls_plan[y][x-1:x+2].count("@") + rolls_plan[y+1][x-1:x+2].count("@"))<=4)&(rolls_plan[y][x]=="@"):
                counter +=1
                rolls_plan[y][x] = "."
    # print(counter)
    
counter
```
