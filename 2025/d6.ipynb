```python
with open("d6.txt", 'r') as f:
    data = f.read()
```

```python
worksheet = [[v for v in l.split(" ") if v!=""] for l in data.split("\n")]
```

```python
total = 0
nb_of_lines = len(worksheet)
for c in range(len(worksheet[0])):
    if worksheet[-1][c] == "+":
        total += sum([int(worksheet[i][c]) for i in range(nb_of_lines-1)])
    elif worksheet[-1][c] == "*":
        total += reduce(lambda x, y: x * y, [int(worksheet[i][c]) for i in range(nb_of_lines-1)], 1)
total
```
