```python
with open('d12.txt', 'r') as f:
    data = f.read()
```

```python
shapes = {int(v.split("\n")[0][:-1]):v.split("\n")[1:] for v in data.split("\n\n")[:-1]}
grids = [([int(i) for i in v.split(": ")[0].split("x")], [int(i) for i in v.split(": ")[1].split(" ")]) for v in data.split("\n\n")[-1].split("\n")]
```

```python
shapes_d = [i/9 for i in [5, 7, 7, 7, 7, 6]]
shapes_d
```

```python
ls = []
lsd = []
for g in grids:
    grid_block_density = sum(g[1])*9/(g[0][0]*g[0][1])
    grid_block_density_super = sum([o*d for o, d in zip(g[1], shapes_d)])*9/(g[0][0]*g[0][1])
    ls.append(grid_block_density)
    lsd.append(grid_block_density_super)
```
