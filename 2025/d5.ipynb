```python
with open("d5.txt", 'r') as f:
    data = f.read()
```

```python
fresh_ranges = [[int(d) for d in v.split("-")] for v in
data.split("\n\n")[0].split("\n")]
ingredient_ids = [int(v) for v in data.split("\n\n")[1].split("\n")]
```

```python
counter = 0
for i in ingredient_ids:
    for fr in fresh_ranges:
        # print(i, fr)
        if i in range(fr[0], fr[1]+1, 1):
            counter +=1
            # print("in")
            break
counter
```

```python

############

# all_ids = set()
all_list = []
checked = [fresh_ranges[0]]
fresh_ranges = fresh_ranges[1:]
# for n, fr in enumerate(fresh_ranges[1:]):
while len(fresh_ranges)>0:
    final_range = fresh_ranges[0]
    fr = final_range
    for c in checked:
        # print(c, final_range)
        if final_range=="EMPTY":
            # print("break")
            break
        elif c == "EMPTY":
            continue
        elif (final_range[0]>=c[0]) & (final_range[1]<=c[1]):
            # print("full overlapp")
            final_range = "EMPTY"
        elif (final_range[0]<c[0]) & (final_range[1]>c[1]):
            # print("Over overlapp")
            fresh_ranges.append([c[1]+1, final_range[1]])
            final_range = [final_range[0], c[0]-1]
        elif (final_range[0] > c[1]) | (final_range[1] < c[0]):
            # print("full avoided")
            continue
        elif final_range[0] >= c[0]:
            # print("final_range[0] >= c[0]")
            final_range = [c[1]+1, final_range[1]]
        elif final_range[0] <= c[0]:
            # print("final_range[0] <= c[0]")
            final_range = [final_range[0], c[0]-1]
        else:
            print(c, final_range)
            raise final_range
    print("initial ", fr, "final", final_range, "" if fr==final_range else "diff")
    checked.append(final_range)
    fresh_ranges = fresh_ranges[1:]
print(sum([c[1]-c[0]+1 if c !="EMPTY" else 0 for c in checked]))

```
